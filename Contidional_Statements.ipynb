```python
#Write a program to categorize a person’s age as a child (0-12), teenager (13-19), or adult (20+).
age=int(input('Enter age:'))
if 0<=age<=12:
       print('child')
elif 13<=age<=19:
       print('teenager')
elif age>=20:
       print('adult')
else:
      print('not valid age')
```
