```python
College = 'Codegnan'

def Student(name, marks):
    original_fee = 50000

    if marks > 90:
        print('Grade A')
        discount = (20 / 100) * original_fee
        fee = original_fee - discount
        print(f'Your fee: {fee}')

    elif marks > 80:
        print('Grade B')
        discount = (15 / 100) * original_fee
        fee = original_fee - discount
        print(f'Your fee: {fee}')

    elif marks > 70:
        print('Grade C')
        discount = (10 / 100) * original_fee
        fee = original_fee - discount
        print(f'Your fee: {fee}')

    else:
        print(f'Your fee: {original_fee}')

name = input('Enter Name: ')
marks = int(input('Enter Marks: '))

Student(name, marks)
```

```python
#*argument
def Order(table,*items,bill):#*argument way
    print(table)
    print(*items)
    print(bill)


Order(15,"friedrice","mutton","chicken","prawns",bill=345)
Order(4,"waterbottle",bill=23)


def Order(table,item1="",item2="",item3="",item4=""):#default way
    print(table)
    print(item1,item2,item3,item4)


Order(3,"chicken fry","egg puff","popcorn","sanwich")
```
