```python
#Print numbers from 1 to 10 using a for loop.
n = int(input()) #n=10
for i in range(n + 1):
  print(i)

```

```python
#Print numbers from 10 to 1 using a while loop.
n=int(input()) #n=10
i=10
while i>0:
    print(i)
    i-=1
```

```python
#Print all even numbers from 1 to 20.
n=int(input()) #n=21
for i in range(1,n+1):
    if i%2==0:
        print(i)
```

```python
#Print the sum of the first 10 natural numbers.
n=int(input())
r=0
for i in range(1,n+1):
    r=r+i
print(r)
```

```python
#Print the multiplication table of a given number.
n=int(input())
for i in range(1,11):
    print(n,'x',i,'=',n*i)
```

```python
#Find the factorial of a given number using a while loop.
n=int(input())
i=1
fact=1
while i<=n:
    fact=fact*i
    i+=1
print(fact)
```

```python
#Check if a number is prime or not.
n = int(input())

i = 2
while i < n:
    if n % i == 0:
        print("not prime")
        break
    i += 1
else:
    print("prime")

```

```python
#Find the sum of digits of a given number.
n=list(map(int,input()))
r=len(n)
sum=0
for i in range(r):
    sum=sum+n[i]
print(sum)
```

```python
#Print all odd numbers from 1 to 50.
n=int(input())
for i in range(1,n+1):
    if i%2!=0:
       print(i)
```

```python
#Find the largest number in a list using a loop.
l=list(map(int,input().split(",")))
r=len(l)
largest=l[0]
for i in range(r):
    if l[i]>largest:
        largest=l[i]
print(largest)
```

```python
#Print all numbers from 1 to 100 that are divisible by 5.
n=int(input())
i=1
while i<=n:
    if i%5==0:
        print(i)
    i+=1
```
