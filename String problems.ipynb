```python
#Ensure a password contains at least 8 characters, a number, and a special character.
password = input("Enter password: ")

has_number = False
has_special = False

for ch in password:
    if ch.isdigit():
        has_number = True
    elif not ch.isalnum():
        has_special = True

if len(password) >= 8 and has_number and has_special:
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
#Count Vowels in a String
s=input()
vowels='aeiouAEIOU'
count=0
for ch in s:
  if ch in vowels:
    count+=1
print(count)
```

```python
#Count consonants in a string
s=input()
vowels='aeiouAEIOU'
count=0
for ch in s:
  if ch not in vowels:
    count+=1
print(count)
```
