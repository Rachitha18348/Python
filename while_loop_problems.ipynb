```python
l=[10,20,30,40,50]
i=0
while i<len(l):
  print(i,l[i])
  i+=1
```

```python
x=['a','b','c','d','e','f']
for i in range(len(x)):
  if i%2==0:
    print(x[i].upper())
```

```python
x=['a','b','c','d','e','f']
i=0
while i<len(x):
  if i%2==0:
    print(x[i].upper())
  else:
    print(x[i].lower())
  i+=1
```

```python
x=['a','b','c','d','e','f']
n=len(x)
i=n-1
while i>=0:
  if i%2==0:
    print(x[i].upper())
  else:
    print(x[i].lower())
  i-=1
```

```python
n=int(input())
for i in range(15):
  if i==n:
    break # to break a condition in loop
  else:
    print(i)
else:
  print('loop completed')
```

```python
n=int(input())
for i in range(15):
  if i==n:
    continue #to avoid particular iteration so 10 got not printed
  else:
    print(i)
else:
  print('loop completed')
```

```python
#nested loop using while
i=0
while i<3:
  j=0
  while j<3:
    print(j)
    j+=1
  i+=1
```

```python
i=0
j=0
while i<=3:
    while j<=3:
        i+=1
        print(j)
        j+=1
```
